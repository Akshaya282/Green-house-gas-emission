```python
import pandas as pd
import plotly.express as px
file_path=r"C:\Users\AKSHAYA\Downloads\SupplyChainEmissionFactorsforUSIndustriesCommodities.xlsx"
excel_file = pd.ExcelFile(file_path)
sheet_names = excel_file.sheet_names
sheet_names
```

```python
data_sheets = [name for name in sheet_names if "Summary" in name or "Detail" in name]
data_sheets
```

```python
def clean_data(df):
    df = df.drop_duplicates()
    df = df.dropna(how='all')
    df = df.dropna(axis=1, how='all')
    return df
```

```python
cleaned_data = {}

for sheet in data_sheets:
    df = pd.read_excel(file_path, sheet_name=sheet)
    df = clean_data(df)
    cleaned_data[sheet] = df
    print(f"{sheet} → {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
for sheet, df in cleaned_data.items():
    numeric_cols = df.select_dtypes(include='number').columns
    text_cols = df.select_dtypes(include='object').columns

    print(f"\n📊 {sheet} preview:")
    display(df.head())  # 👈 Shows top rows of each sheet
    
    if not numeric_cols.empty and not text_cols.empty:
        x = text_cols[0]
        y = numeric_cols[0]

        print(f"Creating chart using: X = {x}, Y = {y}")
        fig = px.bar(df, x=x, y=y, title=f"{sheet} - {y} by {x}")
        fig.show()
    else:
        print("⚠️ Skipping - Not enough data to plot.")

```

```python
def combine_sheets(sheet_list):
    combined_df = pd.DataFrame()
    
    for sheet in sheet_list:
        print(f"Reading: {sheet}")
        df = pd.read_excel(file_path, sheet_name=sheet)
        df = clean_data(df)
        
        year = sheet.split('_')[0]
        df['Year'] = int(year)
        
        combined_df = pd.concat([combined_df, df], ignore_index=True)
        print(f"→ {df.shape[0]} rows added from {sheet}")
    
    print("✅ Combine complete.\n")
    return combined_df

```
